# format strings in plot function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

#b means blue , + is intersection point, -- (dsah repsentation)
plt.plot(x,y,'b+--')
```

```python
#similat to above output
plt.plot(x,y,color='blue', marker='+',linestyle='--')
```

# markersize

```python
plt.plot(x,y,color='blue', marker='+',linestyle='--',markersize=10)
```

# alpha property to control transparency of line chart


```python
#alpha lies between scale 0 to 1
plt.plot(x,y,color='blue', marker='+',linestyle='--',markersize=10,alpha=0.2)
```

```python
plt.plot(x,y,color='blue', marker='+',linestyle='--',markersize=10,alpha=0.8)
```
