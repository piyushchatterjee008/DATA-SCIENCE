```python
%matplotlib inline
import matplotlib.pyplot as plt
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
#Plot a pie chart 
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

# Show percentages for every pie. Specify radius to increase chart size


```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True,autopct='%0.2f%%',radius=.5)
plt.show()
```

# Explode

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True,autopct='%0.0f%%', radius=1.5, explode=[0.2,0.4,0,0,0])
plt.show()
```

# counterclock and angle properties


```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2], startangle=90)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2], startangle=90,center=(2,2))
plt.show()
```
