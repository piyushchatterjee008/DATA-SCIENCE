```python
import pandas as pd
import requests


url = 'https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv'
res = requests.get(url, allow_redirects=True)
with open('weather_data.csv','wb') as file:
    file.write(res.content)
    
    
#Convert string column into the date type   

df = pd.read_csv("weather_data.csv", parse_dates=["day"])
df
```

```python
#use date as an index of dataframe usine set_index() method 
df.set_index('day',inplace=True)
df
```

```python
#Use fillna() method in dataframe
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'no event'
    })
new_df
```

```python
#Use fillna(method="ffill") method in dataframe, forward fill (carry forward the value to null value)

#copying from top
new_df=df.fillna(method='ffill')
new_df
```

```python
#copying from next value to NULL
#Use fillna(method="bfill") method in dataframe

new_df=df.fillna(method='bfill')
new_df
```

```python
#"axis" parameter in fillna() method in dataframe
# axis is either "index" or "columns"
# copy the data right to left columns wise

new_df=df.fillna(method='bfill', axis='columns' )
new_df
```

```python
new_df=df.fillna(method='bfill', axis='index' )
new_df
```

```python
#"limit" parameter in fillna() method in dataframe
# it limits the copying value
new_df=df.fillna(method='bfill', limit=1 )
new_df
```

```python
new_df=df.fillna(method='ffill', limit=2 )
new_df
```

```python
#interpolate() to do interpolation in dataframe, interpolate() method "time

#use linear interpolate to fill the missing number
```

```python
new_df = df.interpolate() 
new_df
```

```python
df.set_index('day',inplace=True)
df
new_df = df.interpolate(method="time") 
new_df
```

# Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)

```python
#dropna() method Drop all the rows which has "na" in dataframe
new_df = df.dropna()
#new_df
```

```python
#"how" parameter in dropna() method 
# drops entire row the were all details are null
new_df = df.dropna(how='all')
new_df
```

```python
#"thresh" parameter in dropna() method 
#thresh=1, minimum 1 valid value should be present in row
##thresh=2, minimum 2 valid value should be present in row

new_df = df.dropna(thresh=1)
new_df
```

```python
##thresh=2, minimum 2 valid value should be present in row
new_df = df.dropna(thresh=2)
new_df
```
