```python
#Which season consisted of the highest number of matches ever played?

sns.set_theme(style="darkgrid")


sns.countplot(x='Season',data=df,palette="Set3")
plt.show()
```

```python
#How to prevent overlapping x-axis labels in sns.countplot

#1.Change figure size


plt.figure(figsize=(12,4)) # this creates a figure 8 inch wide, 4 inch high
sns.countplot(x="Season", data=df)
plt.show()
```

```python
#2.Rotate the ticklabels

ax = sns.countplot(x="Season", data=df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40)
plt.tight_layout()
plt.show()
```

```python
#Decrease Fontsize

ax = sns.countplot(x="Season", data=df)

ax.set_xticklabels(ax.get_xticklabels(), fontsize=7)
plt.tight_layout()
plt.show()
```
